```python
#Location of Restaurant using longitude & Latitude
import folium

data = pd.read_csv('Restaurant.csv')
m = folium.Map(location=[20.5937, 78.9629], zoom_start=5)

for _, row in data.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']]).add_to(m)

m.save("restaurant_map.html")

```

```python
import folium
from folium.plugins import MarkerCluster

data = pd.read_csv('Restaurant.csv')

m = folium.Map(location=[20.5937, 78.9629], zoom_start=5)
marker_cluster = MarkerCluster().add_to(m)

# Restaurant locations
for idx, row in data.iterrows():
    if not pd.isnull(row['Latitude']) and not pd.isnull(row['Longitude']):
        folium.Marker(
            location=[row['Latitude'], row['Longitude']],
            popup=row['Restaurant Name']
        ).add_to(marker_cluster)

# Show map
m.save("location.html")
print("Map saved as location.html")
```

```python
import pandas as pd
import plotly.express as px

# Load data
data = pd.read_csv("Restaurant.csv")

# Ensure correct data types
data['Longitude'] = pd.to_numeric(data['Longitude'], errors='coerce')
data['Latitude'] = pd.to_numeric(data['Latitude'], errors='coerce')
data['Aggregate rating'] = pd.to_numeric(data['Aggregate rating'], errors='coerce')

# Drop rows with missing values in essential columns
data = data.dropna(subset=['Longitude', 'Latitude', 'Aggregate rating', 'Restaurant Name'])

# 3D Scatter Plot
fig = px.scatter_3d(
    data,
    x='Longitude',
    y='Latitude',
    z='Aggregate rating',
    color='Aggregate rating',
    hover_name='Restaurant Name',
    title='3D Plot of Restaurant Ratings by Location',
    color_continuous_scale='Viridis'
)

fig.write_html("restaurant_3d_plot.html")
```

```python
# Table Booking
table_booking_counts = df['Has Table booking'].value_counts(normalize=True) * 100

# Online Delivery
online_delivery_counts = df['Has Online delivery'].value_counts(normalize=True) * 100

print("Table Booking %:\n", table_booking_counts)
print("\nOnline Delivery %:\n", online_delivery_counts)
```

```python
import matplotlib.pyplot as plt

# Plot Pie Charts
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
table_booking_counts.plot.pie(autopct='%1.1f%%', title='Table Booking', colors=['red', 'black'])

plt.subplot(1, 2, 2)
online_delivery_counts.plot.pie(autopct='%1.1f%%', title='Online Delivery', colors=['red', 'black'])

plt.tight_layout()
plt.show()
```

```python

rating_comparison = df.groupby('Has Table booking')['Aggregate rating'].mean().reset_index()  # Group by Table Booking and get average ratings

print(rating_comparison)
```

```python
import seaborn as sns

sns.barplot(data=rating_comparison, x='Has Table booking', y='Aggregate rating', palette=['red', 'black'])
plt.title("Avg Ratings: Table Booking vs No Booking")
plt.ylabel("Average Rating")
plt.xlabel("Table Booking")
plt.show()
```

```python
delivery_vs_price = df.groupby(['Price range', 'Has Online delivery']).size().unstack().fillna(0)

# Convert to percentages
delivery_percent = (delivery_vs_price.T / delivery_vs_price.sum(axis=1)).T * 100

print(delivery_percent)
```

```python
delivery_percent.plot(kind='bar', stacked=True, color=['red', 'black'])
plt.title("Online Delivery Availability by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Percentage")
plt.legend(title="Online Delivery")
plt.xticks(rotation=0)
plt.show()
```
